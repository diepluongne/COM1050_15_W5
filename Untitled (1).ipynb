# A3

```python
def nguyento(a):
    if a<=1:
        return "NO"
    for i in range (2,int(a**0.5)+1):
        if a%i==0:
            return "NO"
    return "YES"
n=int(input())
nguyento(n)
```

```python
n=int(input())
def sohoanhao(a):
    tong=0
    for i in range (1,a):
        if a%i==0:
            tong=tong+i
    if tong==a:
        return "YES"
    else:
        return "NO"
sohoanhao(n)
        
```

```python
def cal(a,b,c):
    if c=="+":
        return a+b
    elif c=="-":
        return a-b
    elif c=="*":
        return a*b
    elif c=="/":
        return a/b
n1,n2,oper=input().split()
n1=float(n1)
n2=float(n2)
ketqua=cal(n1,n2,oper)
round(ketqua,2)
```
